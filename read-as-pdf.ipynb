```python
import PyPDF2
from pdfminer.high_level import extract_pages
from pdfminer.layout import LTTextContainer, LTChar
```

```python
def text_extraction(elem):
    text = elem.get_text()
    formats = []
    for line in elem:
        if isinstance(line, LTTextContainer):
            for char in line:
                if isinstance(char, LTChar):
                    formats.append(char.fontname)
                    formats.append(char.size)
    format_line = list(set(formats))
    return text, format_line
```

```python
path = '/kaggle/input/book-pdf/1. Rebilius Crs (Robinson Crusoe) Autor Daniel Defoe.pdf'
pdf_file = open(path, 'rb')
pdf_readed = PyPDF2.PdfReader(pdf_file)
page_text_dict = {}

for num, page in enumerate(extract_pages(path)):
    page_text = []
    line_format = []
    page_content = []
    
    page_elements = [(element.y1, element) for element in page._objs]
    page_elements.sort(key=lambda a: a[0], reverse=True)

    for i,component in enumerate(page_elements):
        element = component[1]
        if isinstance(element, LTTextContainer):
            (line_text, format_per_line) = text_extraction(element)
            line_format.append(format_per_line)
            page_content.append(line_text)
        else:
            pass

    page_text_dict['Page_'+str(num)] = [page_text, line_format, page_content]
    
pdf_file.close()
```
